```python
import pandas as pd

df_energy = pd.read_csv('global energy consumption.csv')
for i in range(3, 12):
    df_energy.iloc[:, i+1] += df_energy.iloc[:, i]
df_energy
```

```python
df_energy.columns
```

```python
import plotly.graph_objs as go

fig = go.Figure()
fig.add_trace(go.Scatter(
name = 'Traditional biomass',
x = df_energy['Year'],
y = df_energy['Traditional biomass'],
fill = 'tozeroy',
mode = 'lines'))

for en in df_energy.columns[4:]:
    fig.add_trace(go.Scatter(
    name = en,
    x = df_energy['Year'],
    y = df_energy[en],
    fill = 'tonexty',
    mode = 'lines'))
    
fig.update_layout(title = 'Global primary energy consumption by source',
                  yaxis=dict(range=[0, 190000]),
                  paper_bgcolor='darkgray',
                  plot_bgcolor='darkgray')

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Sample data for two intersecting lines
x = np.linspace(0, 10, 100)
y1 = 2 * x
y2 = -x + 10

# Find the intersection point (x, y) where y1 = y2
intersection_x = (10 - 0) / (2 + 1)  # Intersection x-coordinate
intersection_y = 2 * intersection_x  # Intersection y-coordinate

# Create a Plotly figure with the two lines
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='Line 1'))
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Line 2'))

# Add a shape to highlight the intersection point
fig.add_shape(type='circle',
              x0=intersection_x - 0.5, y0=intersection_y - 0.5,
              x1=intersection_x + 0.5, y1=intersection_y + 0.5,
              line=dict(color='red', width=2),
              fillcolor='rgba(255, 0, 0, 0.5)')

# Add an annotation to label the intersection point
fig.add_annotation(x=intersection_x, y=intersection_y,
                   text='Intersection',
                   showarrow=True,
                   xshift=0, yshift=30,
                   bgcolor='white',
                   bordercolor='red',
                   borderwidth=2,
                   opacity=0.8)

# Update layout for better visualization
fig.update_layout(
    title='Intersection of Two Lines',
    xaxis_title='X',
    yaxis_title='Y'
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Create the DataFrame
df = pd.DataFrame({
    'Year': [1960, 1970, 1980, 1990, 2000],
    'TB': [50, 100, 150, 200, 250],
    'Coal': [0, 0, 100, 200, 300],
    'Oil': [0, 0, 0, 300, 600]
})

# Plot the line graphs for each column (excluding 'Year')
fig = go.Figure()

for column in df.columns[1:]:  # Exclude the 'Year' column
    fig.add_trace(go.Scatter(x=df['Year'], y=df[column], mode='lines', name=column))

# Determine intersection points
intersection_points = []
for i in range(len(df) - 1):
    for j in range(i + 1, len(df)):
        year1, year2 = df.loc[i, 'Year'], df.loc[j, 'Year']
        for column in df.columns[1:]:
            value1, value2 = df.loc[i, column], df.loc[j, column]
            if (value1 < value2 and i < len(df) - 2 and value2 < df.loc[i + 1, column]) \
                    or (value1 > value2 and j < len(df) - 1 and value1 < df.loc[j + 1, column]):
                intersection_points.append((year1 + year2) / 2)

# Add annotations at intersection points
for year in intersection_points:
    for column in df.columns[1:]:
        value = df.loc[df['Year'] == year, column].iloc[0]
        fig.add_annotation(x=year, y=value,
                            text=f'{column} Intersection\nYear: {year}, Value: {value}',
                            showarrow=True,
                            arrowhead=1,
                            ax=0,
                            ay=-40)

# Update layout
fig.update_layout(title='Line Graph with Intersection Annotations',
                  xaxis_title='Year',
                  yaxis_title='Value')

# Show the plot
fig.show()
```
